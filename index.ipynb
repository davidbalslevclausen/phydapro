# Example w binder

```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt


%matplotlib inline
```

# Amplitude modulation

```python
N = 10000
f = 42
A = 50.
fA = 1.5
x = np.linspace(0,30,N)
y = A*(1+0.5*np.sin(2*np.pi*fA*x))* np.sin(2*np.pi*f*x)

fig, (ax,ax2) = plt.subplots(2,1,figsize=(8,6))
ax.plot(x,y)
ax.set_xlim(0,4)

w = np.fft.rfft(y)
k = np.fft.rfftfreq(len(x),30/N)
ax2.plot(k,np.abs(w))
ax2.set_xlim(37,47)
```

```python
N = 50000
f = 40
A = 50.
fA = 0
fw = 0.5
fwA = 0.2
x = np.linspace(0,30,N)
y = A*(1+0.5*np.sin(2*np.pi*fA*x))* \
         np.sin(2*np.pi*(f*x + fwA*np.sin(2*np.pi*fw*x)))

fig, (ax,ax2) = plt.subplots(2,1,figsize=(8,6))
ax.plot(x,y)
ax.set_xlim(0,2)

w = np.fft.rfft(y)
k = np.fft.rfftfreq(len(x),30/N)
ax2.plot(k,np.abs(w))
ax2.set_xlim(25,55)
```
